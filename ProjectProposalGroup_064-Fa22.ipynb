# Research Question

How does the rising ocean temperatures and sea level affect the frequency of unprovoked shark attacks in the Northern Hemisphere?

## Background and Prior Work

Sharks are one of the oceans most notable charismatic megafauna. Whether you are a 10 year old who thinks sharks are the coolest or a 22 year old who is scared to death by the thought of them, sharks are a species that everyone knows. This is due to Popular forms of media such as movies and news outlets that make great commotion when there's a shark attack or sighting making it seem that every time you go in the ocean you will be pounced on by one of these apex predators. In reality the stereotype does not hold and shark attacks are fairly rare. However, to many beach goers, the likelihood of encountering a shark or being attacked by one could be increased due to climate change bringing rising tempuratures.
 
This topic has been increasingly talked about in the past year especially with more citings off the coast of California and even San Diego. Many local news outlets have also made reports talking about the correlation CBS8 quoting Dr. Lowe of the Cal State Long Beach Shark Lab in the recent story *Why are sharks staying around the California coast longer?* saying "“We know that climate change is starting to influence the behavior and distribution of the sharks"[^Hope]. News stories like this put the topic on our group's radar, especially since we are based out of San Diego, making us think that approaching the change in distribution of sharks compared to the change in temperatures due to climate change could be interesting when looked through the lens of a data scientist.
 
Finally, going deeper into more sophisticated research we find that many high ups in academia are also pondering the implications of climate change on these apex predators. M. Schlaff et al. in *Influence of environmental factors on shark and ray movement, behavior and habitat use: a review* says

>"A recent integrated risk assessment looking at the vulnerability of sharks and rays on Australia’s Great Barrier Reef (GBR) to climate change found that vulnerability was driven by case-specific interactions of multiple factors (e.g. water temperature, ocean acidification) and species attributes (e.g. habitat specificity, mobility) (Chin et al. 2010). Based on this, the authors concluded that freshwater/estuarine and reef associated sharks and rays are particularly vulnerable to climate change impacts since key habitats for these species are affected by an array of anthropogenic and environmental impacts, thus reducing the overall resilience of these species"[^Schlaff]

Aswell Hammerschlag et al. in *Ocean warming alters the distributional range, migratory timing, and spatial protections of an apex predator, the tiger shark (Galeocerdo cuvier)* says 

>"Given climate change threats to ecosystems, it is critical to understand the responses of species to warming. This is especially important in the case of apex predators since they exhibit relatively high extinction risk, and changes to their distribution could impact predator–prey interactions that can initiate trophic cascades"[^Hammerschlag]

Both these papers lend insight into the issues that climate change is affecting shark range and distbution but both are aprocahing the issue from a much more ecolocigal point of view loking into how climate change is effecting the overall ecosystems that these predators inhabit. Our gorup however is taking a different approach then these scientific minds and will be looking with data science techniques into how shark to human interactions are changing with a warming in tempurate due to climate change.

# Ethics & Privacy

The dataset we used may have issues regarding ethics, as the names of the victims of shark attacks are listed. However, confidential data such as ones name, age, gender, and the exact location the attack ocurred will be omitted from our own dataset, as it is not needed for our research question. Our only concerns are the date of the shark attack, the country in which it occurred, and what caused the attack. Because our data focuses on shark attacks, an animal that lives in the ocean, we do expect a bias towards countries located next to oceans. For example, we cannot expect to see a shark attack in Austria or Hungary, as they are surrounded by neighboring countries, away from the ocean. Yet, these biases should not be problematic as these countries have no relevance to our question. Datasets concerning ocean temperatures should also not pose any ethical concerns or biases as our data is only concerned with temperatures in the Northern Hemisphere. There will be no violations of ones personally identifiable information (PII) when analyzing trends in ocean temperature. As for sea level, we believe there are no ethical violations, as no PII can be retrieved from the datasets concerning sea level changes throughout the years.
